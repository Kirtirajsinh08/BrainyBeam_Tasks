```python
# We are given an employee data having following columns:
# Employee_Name
# Department
# Daily_Sales
# Date

# Now we will implement function groupby and apply so that we can get to know who is the best performer of a particular day 
# from a particular department
```

```python
import pandas as pd
import numpy as np
```

```python
employee_data = {
    'Employee_Name': ['Sine', 'Cos', 'Theta', 'Sine', 'Cos', 'Theta', 'Gamma', 'Beta', 'Gamma', 'Beta'],
    'Department': ['Sales', 'Sales', 'Sales', 'Sales', 'Sales', 'Sales', 'HR', 'HR', 'HR', 'HR'],
    'Daily_Sales': [200, 300, 250, 150, 100, 300, 250, 150, 100, 150],
    'Date': ['2024-12-10', '2024-12-10', '2024-12-10', '2024-12-11', '2024-12-11', '2024-12-11', '2024-12-10', '2024-12-10', '2024-12-11', '2024-12-11']
}
df = pd.DataFrame(employee_data)
df = df.reset_index(drop=True)
df
```

```python
def employee_rank(data):
    data['Performance_Rank'] = data['Daily_Sales'].rank(ascending=False, method='min')
    return data

new_data = df.groupby(['Department', 'Date']).apply(employee_rank)
new_data
```

```python
new_df = new_data.drop(columns=['Department','Date'])
new_df
```

# Working on a CSV File

```python
# Getting insights of a Particular Region using "groupby" and "apply"
```

```python
data_set = pd.read_csv("100_Sales.csv")
data_set.sample(5)
```

```python
data_set = data_set.drop(columns=["Unnamed: 9", "Unnamed: 10"], axis=1)
data_set.sample(5)
```

```python
def region_wise_categorisation(data):
    
    insights = {
        "Average_Unit_Cost" : data["Unit_Cost"].mean(),
        "Total_Revenue" : data["Total_Revenue"].sum(),
        "Total_Profit" : data["Total_Profit"].sum(),
        "Top_Country_Making_Profit" : data.loc[data["Total_Profit"].idxmax(), "Country"]
    }
    
    return pd.Series(insights)

region_insight = data_set.groupby(["Region"]).apply(region_wise_categorisation)
region_insight
```
